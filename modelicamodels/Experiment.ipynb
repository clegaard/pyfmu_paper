```python
def simulate_plot(width):
    bike = BikeDynamicModel()

    bike.a = lambda t: 0.0
    bike.deltaf = Driver.get_steering(width)

    bike.vx0 = 1.0

    bike.simulate(100, 0.01)
    
    _, axs = plt.subplots(3)
    
    axs[0].plot(bike.X, bike.Y, label='X vs Y')
    axs[1].plot(bike.ts, bike.ar, label='ar')
    axs[2].plot(bike.ts, bike.psi, label='psi')
    plt.show()
```

```python
simulate_plot(3.0)
```

```python
simulate_plot(30.0)
```
